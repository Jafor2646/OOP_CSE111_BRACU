```python
#Task 1
class Student:
  id = 0
  def __init__(self, n, d, a, c):
    self.id = 0
    self.name = n
    self.dept = d
    self.age = a
    self.cgpa = c
    Student.id += 1
  def get_details(self):
    print("ID: "+ str(Student.id)+"\nName: "+self.name+"\nDepartment: "+self.dept+"\nAge: "+str(self.age)+"\nCGPA: "+str(self.cgpa))
  @classmethod
  def from_String(cls, s):
    n, d, a, c = s.split("-")
    return cls(n, d, int(a), float(c))
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
```

```python
#Task 2
class Assassin:
  count = 0
  def __init__(self, n, p):
    Assassin.count  += 1
    self.name = n
    self.success = p
  def printDetails(self):
    print("Name:",self.name)
    print("Success rate: ",self.success,"%",sep='')
    print("Total number of Assassin:", Assassin.count)
  @classmethod
  def failureRate(cls, n , f):
    return cls(n, 100-f)
  @classmethod
  def failurePercentage(cls, n, f):
    return cls(n, (100 - int(f[:-1])))

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#Task 3
class Passenger:
  count = 0
  def __init__(self, n):
    self.name = n
    self.fare = 450
    Passenger.count += 1
  def set_bag_weight(self, w):
    if w > 50:
      self.fare += 100
    elif w > 20 and w <= 50:
      self.fare += 50 
  def printDetail(self):
    print("Name:",self.name)
    print("Bus Fare:", self.fare)
print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#Task 4
class Travel:
  count = 0
  def __init__(self, s , d):
    Travel.count += 1
    self.source = s
    self.des = d
    self.time = "1:00"
  def display_travel_info(self):
    temp = "Source: "+self.source + "\nDestination:" + self.des + "\nFlight Time:" + self.time
    return temp
  def set_time(self, t):
    self.time = str(t) + ":00"
  def set_source(self, s):
    self.source = s
  def set_destination(self, d):
    self.des = d
print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#Task 5
class Employee:
  def __init__(self, n, w):
    self.name = n
    self.workingPeriod = w
  @classmethod
  def employeeByJoiningYear(cls, n, y):
    name = n
    work = 2022 - y
    return cls(name, work)
  @staticmethod
  def experienceCheck(e, g):
    if e < 3:
      if g.lower() == "male":
        return "He is not experienced"
      else:
        return "She is not experienced"
    else:
      if g.lower() == "male":
        return "He is experienced"
      else:
        return "She is experienced"
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 7
class Cat:
  Number_of_cats = 0
  def __init__(self, c, d):
    self.color = c
    self.todo = d
    Cat.Number_of_cats += 1
  @classmethod
  def no_parameter(cls):
    return cls("White", "sitting")
  @classmethod
  def first_parameter(cls, c):
    return cls(c, "sitting")
  @classmethod
  def second_parameter(cls, d):
    return cls("Grey", d)
  def printCat(self):
    print(self.color, "cat is", self.todo)
  def changeColor(self, c):
    self.color = c
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8
import math
class Cylinder:
  radius = 5
  height = 18
  def __init__(self, r, h):
    print("Default radius="+str(Cylinder.radius)+" and height="+str(Cylinder.height)+".")
    Cylinder.radius = r
    Cylinder.height = h
    print("Updated:", "radius="+str(Cylinder.radius),"and","height="+str(Cylinder.height)+".")
  @staticmethod
  def area(r, h):
    total = (2*math.pi*(r**2)) + (2*math.pi*r*h)
    print("Area:", float(total))
  @staticmethod
  def volume(r, h):
    total = math.pi*(r**2)*h
    print("Volume:", total)
  @classmethod
  def swap(cls, h, r):
    return cls(r, h)
  @classmethod
  def changeFormat(cls, s):
    rad , hight = s.split("-")
    return cls(float(rad), float(hight))
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10
class SultansDine:
  branch = 0
  sell = 0
  obj = []
  def __init__(self, n):
    self.branchName = n
    self.branchSell = 0
    SultansDine.branch += 1
    SultansDine.obj.append(self)
  def sellQuantity(self, q):
    if q < 10:
      self.branchSell += (q*300)
    elif q < 20:
      self.branchSell += (q*350)
    else:
      self.branchSell += (q*400)
    SultansDine.sell += self.branchSell
  def branchInformation(self):
    print("Branch Name:", self.branchName)
    print("Branch Sell:", self.branchSell,"Taka")
  @classmethod
  def details(cls):
    print("Total Number of branch(s):", cls.branch)
    print("Total Sell:",cls.sell,"Taka")
    if len(cls.obj) != 0:
      for i in cls.obj:
        print("Branch Name:", i.branchName +",","Branch sell:",i.branchSell,"Taka")
        percentage = "{:.2f}".format((i.branchSell/cls.sell)* 100)
        print("Branch consists of total sell's:", percentage + "%")
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
