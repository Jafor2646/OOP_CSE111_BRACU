```python
#Dictionary
#Task 1
uniqueValue = []
myList = []
myDict1 = {}
myDict2 = {}
input1 = input("Enter dictionary:").split(", ")
input2 = input("Enter dictionary:").split(", ")
for i in input1:
  k,v = i.split(": ")
  myDict1[k] = int(v)
for i in input2:
  k,v = i.split(": ")
  myDict2[k] = int(v)
for k,v in myDict2.items():
  if k not in myDict1:
    myDict1[k] = v
  else:
    myDict1[k] += v
print(myDict1)
for i in myDict1.values():
  myList.append(i)
for i in range(len(myList)):
  minIndex = i
  for j in range(i+1,len(myList)):
    if myList[j] < myList[minIndex]:
      minIndex = j
  temp = myList[i]
  myList[i] = myList[minIndex]
  myList[minIndex] = temp
for i in myList:
  if i not in uniqueValue:
    uniqueValue.append(i)
print("Values:",tuple(uniqueValue))
```

```python
#Function
#Task 1
def BMI(height,weight):
  meter = height /100
  bmi = weight / (meter**2)
  bmi = str(bmi)[0:4]
  bmi = float(bmi)
  if bmi < 18.5:
    print("Score is",bmi,"You are underwight")
  elif bmi >= 18.5 and bmi <= 24.9:
    print("Score is",bmi,"You are normal")
  elif bmi >= 25 and bmi <= 30:
    print("Score is",bmi,"You are overweight")
  elif bmi > 30:
    print("Score is",bmi,"You are obese")
var = input("Enter value like tuple:")[1:-1].split(", ")
h = int(var[0])
w = int(var[1])
BMI(h,w)
```

```python
#Task 2
def functionName(min,max,divisor):
  sum = 0
  for i in range(min,max):
    if i % divisor == 0:
      sum += i
  return sum
var = input("Enter value like tuple:")[1:-1].split(", ")
print(functionName(int(var[0]),int(var[1]),int(var[2])))
```

```python
#Task 3
def burgerPriceCalculator(burger,place='Mohakhali'):
    if place=='Mohakhali':
        delivary= 40
    else:
        delivary = 60
    if burger == 'BBQ Chicken Cheese Burger':
        burgerPrice= 250
    elif burger == 'Beef Burger':
        burgerPrice= 170
    elif burger == 'Naga Drums':
        burgerPrice= 200
    tax = (burgerPrice/100)*8
    return burgerPrice+delivary+tax
var=input('Enter your order and place:')[1:-1]
if ',' in var:
  burger,place = var.split(", ")
  print(burgerPriceCalculator(burger[1:-1],place[1:-1]))
else:
  print(burgerPriceCalculator(var[1:-1]))
```

```python
#Task 4
def replaceDomain(email,new,old='kaaj.com'):
    for i in range(len(email)):
        if email[i]=='@':
            if email[i+1:] == old:
                newMail = email[:i+1] + new
                change = True
            else:
                newMail = email
                change = False
            break
    if change:
        print('Changed:',newMail)
    else:
        print('Unchanged:',newMail)
st=input('Enter string:')[1:-1].split(',')
if len(st)>2:
    email,new,old=st
    email = email[1:-1]
    replaceDomain(email,new,old)
else:
    email,new=st
    email = email[1:-1]
    replaceDomain(email,new)
```

```python
#Task 5
def palindromeChecker(word):
    wordLenght=len(word)
    secondHalfReverse = ''
    if len(word) % 2 ==0:
        firstHalf=word[:(wordLenght//2)]
        secondHalf=word[(wordLenght//2):]
    else:
        firstHalf=word[:((wordLenght)//2)]
        secondHalf=word[(((wordLenght)//2)+1):]
    for rev in range(len(secondHalf)-1,-1,-1):
        secondHalfReverse += secondHalf[rev]
    if firstHalf == secondHalfReverse:
        print('Palindrome')
    else:
        print('Not a palindrome')
st=input('Enter:')
temp_list = list()
final_word = ''
for i in st:
    if i != ' ':
        temp_list.append(i)
for i in temp_list:
    final_word += i
palindromeChecker(final_word)
```

```python
#Task 6
def grammerCorrector(paragraph):
    myList= []
    special = '.!?'
    newParagraph = ''
    for i in paragraph:
        myList.append(i)
    myList[0]=myList[0].upper()
    for i in range(len(myList)):
        if myList[i] in special and i!= len(myList)-1:
            myList[i+2] = myList[i+2].upper()
        if myList[i] == 'i' and myList[i+1] ==' ' and myList[i-1] == ' ':
            myList[i]=myList[i].upper()
    for i in myList:
        newParagraph += i
    return newParagraph
paragraph=input('Enter your sentence:')[2:-2]
new_paragraph = grammerCorrector(paragraph)
print(new_paragraph)
```
