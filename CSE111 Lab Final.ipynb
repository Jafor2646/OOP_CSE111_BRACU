```python
class GameOfThrones:
    chars = {}
    all_status = {}
    def __init__(self, name, age, title):
        self.__name = name
        self.__age = age
        self._title = title
        self._status = 'Alive or Dead'
    def setName(self, name):
        self.__name = name
    def getName(self):
        return self.__name
    def getAge(self):
        return self.__age
    def killPerson(self):
        print('The Lannisters send their regards!')
# ------------------
class Person(GameOfThrones):
    num = 0
    def __init__(self,name,age,title,house='Stark'):
        super().__init__(name,age,title)
        self.house=house
        self._status = 'Alive'
        Person.num+=1
        GameOfThrones.all_status[self.getName()]=self._status
    @staticmethod
    def showGotInfo():
        print(f'Number of Person: {len(GameOfThrones.all_status)}')
        print(GameOfThrones.all_status)
    def __str__(self):
        return f'Details of the character:\nName: {self.getName()}\nAge: {self.getAge()}\nHouse: {self.house}\nTitle: {self._title}\nStatus: {self._status}'
    def killPerson(self):
        self._status='Dead'
        GameOfThrones.all_status[self.getName()] = self._status
        super().killPerson()
    def hasDragon(self):
        if self.house=='Targaryen':
            return True
        else:
            return False
#------------------
class WhiteWalker(GameOfThrones):
    def __init__(self,name):
        self.name=name
    @classmethod
    def createWhiteWalker(cls,name):
        if name._status=='Alive':
            print(f'No White Walker created since the person is alive!')
        else:
            print(f'Danger! A White Walker is created!')
            return cls(name.getName())

GameOfThrones.chars = {1:Person('Daenerys Targaryen', 25, 'Mother ofDragons','Targaryen')}
GameOfThrones.chars.update({2: Person('Jon Snow', 23, 'King in the North')})
GameOfThrones.chars.update({3: Person('Sansa Stark', 20, 'Queen in the North'), 4:Person('Arya Stark', 16, 'No One')})
Person.showGotInfo()
print(1,'-'*25)
print(GameOfThrones.chars[1])
print(2,'-'*25)
print(GameOfThrones.chars[2])
print(3,'-'*25)
GameOfThrones.chars[1].killPerson()
print(4,'-'*25)
print(GameOfThrones.chars[1])
print(5,'-'*25)
white1 =WhiteWalker.createWhiteWalker(GameOfThrones.chars[1])
print(6,'-'*25)
print("White Walker's name is", white1.name)
print(7,'-'*25)
white2 =WhiteWalker.createWhiteWalker(GameOfThrones.chars[2])
print(8,'-'*25)
print(Person.hasDragon(GameOfThrones.chars[1]))
print(9,'-'*25)
print(Person.hasDragon(GameOfThrones.chars[2]))
print(10,'-'*25)
Person.showGotInfo()


# Number of Person: 4
# All Person Status: {'Daenerys Targaryen': 'Alive', 'Jon Snow': 'Alive', 'Sansa Stark': 'Alive', 'Arya Stark': 'Alive'}
# 1 -------------------------
# Details of the character:
# Name: Daenerys Targaryen
# Age: 25
# House: Targaryen
# Title: Mother of Dragons
# Status: Alive
# 2 -------------------------
# Details of the character:
# Name: Jon Snow
# Age: 23
# House: Stark
# Title: King in the North
# Status: Alive
# 3 -------------------------
# The Lannisters send their regards!
# 4 -------------------------
# Details of the character:
# Name: Daenerys Targaryen
# Age: 25
# House: Targaryen
# Title: Mother of Dragons
# Status: Dead
# 5 -------------------------
# Danger! A White Walker is created!
# 6 -------------------------
# WhiteWalker's name is Daenerys Targaryen
# 7 -------------------------
# No White Walker created since the person is alive!
# 8 -------------------------
# True
# 9 -------------------------
# False
# 10 -------------------------
# Number of Person: 4
# All Person Status: {'Daenerys Targaryen': 'Dead', 'Jon Snow': 'Alive', 'Sansa Stark': 'Alive', 'Arya Stark': 'Alive'}
```
